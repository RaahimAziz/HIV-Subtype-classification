```python
#opening fasta file for HIV-1:
HIV1 = "/content/hiv1.fna"
HIV1_fasta_records = SeqIO.parse(HIV1, "fasta")

#Saving records in a dictionary:
HIV1_fasta = {"ID": None,
              "Name": None,
              "Description": None,
              "Sequence": None}

for record in HIV1_fasta_records:
    HIV1_fasta["ID"] = record.id
    HIV1_fasta["Name"] = record.name
    HIV1_fasta["Description"] = record.description
    HIV1_fasta["Sequence"] = str(record.seq)
```

```python
HIV2 = "/content/hiv2.fna"
HIV2_fasta_records = SeqIO.parse(HIV2, "fasta")

HIV2_fasta = {"ID":None,
              "Name": None,
              "Description": None,
              "Sequence": None}

for record in HIV2_fasta_records:
    HIV2_fasta['Name'] = record.name
    HIV2_fasta['ID'] = record.id
    HIV2_fasta['Description'] = record.description
    HIV2_fasta['Sequence'] = str(record.seq)
```

```python
#Comparing bothe sequences:
HIV11 = HIV1_fasta["Sequence"]
HIV22 = HIV2_fasta["Sequence"]

```

```python
print(f"HIV-1 is {len(HIV11)} basepairs long")
print(f"HIV-2 is {len(HIV22)} basepairs long")

#HIV-1 has fewer number of basepairs as compared to HIV-2
```

```python
Aligner = Align.PairwiseAligner()
alignments = Aligner.align(HIV11,HIV22)
```
